```python

class DoubleConv(nn.Module):
    def __init__(self, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.LazyConv2d(out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.LazyConv2d(out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1, features=[64, 128, 256, 512]):
        super(UNet, self).__init__()
        self.encoder = nn.ModuleList()
        self.decoder = nn.ModuleList()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Encoder
        for feature in features:
            self.encoder.append(DoubleConv(feature))

        # Bottleneck
        self.bottleneck = DoubleConv(features[-1] * 2)

        # Decoder
        for feature in reversed(features):
            self.decoder.append(nn.ConvTranspose2d(feature * 2, feature, kernel_size=2, stride=2))
            self.decoder.append(DoubleConv(feature))  # after concatenation

        self.final_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)

    def forward(self, x):
        skip_connections = []

        for layer in self.encoder:
            x = layer(x)
            skip_connections.append(x)
            x = self.pool(x)

        x = self.bottleneck(x)
        skip_connections = skip_connections[::-1]

        for idx in range(0, len(self.decoder), 2):
            x = self.decoder[idx](x)  # upsample
            skip_connection = skip_connections[idx // 2]
            if x.shape != skip_connection.shape:
                x = F.interpolate(x, size=skip_connection.shape[2:], mode='bilinear', align_corners=True)
            x = torch.cat((skip_connection, x), dim=1)  # concat
            x = self.decoder[idx + 1](x)  # double conv

        return self.final_conv(x)
```

```python
class UNetGenerator(nn.Module):
    def __init__(self, out_channels=3):
        super(UNetGenerator, self).__init__()
        self.encoder = nn.ModuleList([
            nn.Sequential(
                nn.LazyConv2d(64, kernel_size=4, stride=2, padding=1),
                nn.LeakyReLU(0.2)
            ),  # 256x256
            nn.Sequential(
                nn.LazyConv2d(128, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 128x128
            nn.Sequential(
                nn.LazyConv2d(256, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 64x64
            nn.Sequential(
                nn.LazyConv2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 32x32
            nn.Sequential(
                nn.LazyConv2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 16x16
            nn.Sequential(
                nn.LazyConv2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 8x8
            nn.Sequential(
                nn.LazyConv2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.LeakyReLU(0.2)
            ),  # 4x4
            nn.Sequential(
                nn.LazyConv2d(512, kernel_size=4, stride=2, padding=1),
                nn.ReLU()
            )  # 2x2
        ]) 
        self.decoder = nn.ModuleList([
            nn.Sequential(
                nn.LazyConvTranspose2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.Dropout(0.5),
                nn.ReLU()
            ),  # 4x4
            nn.Sequential(
                nn.LazyConvTranspose2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.Dropout(0.5),
                nn.ReLU()
            ),  # 8x8
            nn.Sequential(
                nn.LazyConvTranspose2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.Dropout(0.5),
                nn.ReLU()
            ),  # 16x16
            nn.Sequential(
                nn.LazyConvTranspose2d(512, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.ReLU()
            ),  # 32x32
            nn.Sequential(
                nn.LazyConvTranspose2d(256, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.ReLU()
            ),  # 64x64
            nn.Sequential(
                nn.LazyConvTranspose2d(128, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.ReLU()
            ),  # 128x128
            nn.Sequential(
                nn.LazyConvTranspose2d(64, kernel_size=4, stride=2, padding=1),
                nn.LazyBatchNorm2d(),
                nn.ReLU()
            ),  # 256x256
        ])

        self.final = nn.Sequential(
            nn.LazyConvTranspose2d(out_channels, kernel_size=4, stride=2, padding=1),
            nn.Tanh()  # outputs in range [-1, 1]
        )  # 512x512

    def forward(self, x):
        enc_feats = []
        for enc_layer in self.encoder:
            x = enc_layer(x)
            enc_feats.append(x)

        for dec_layer in self.decoder:
            x = dec_layer(x)
            skip_feat = enc_feats.pop(-2)  # skip connection
            x = torch.cat([x, skip_feat], dim=1)

        x = self.final(x)
        return x

class PatchDiscriminator(nn.Module):
    def __init__(self, in_channels=6):  # real+fake or input+generated pair
        super(PatchDiscriminator, self).__init__()
        self.model = nn.Sequential(
            nn.LazyConv2d(64, kernel_size=4, stride=2, padding=1),  # 256
            nn.LeakyReLU(0.2),
            nn.LazyConv2d(128, kernel_size=4, stride=2, padding=1),
            nn.LazyBatchNorm2d(),
            nn.LeakyReLU(0.2),
            nn.LazyConv2d(256, kernel_size=4, stride=2, padding=1),
            nn.LazyBatchNorm2d(),
            nn.LeakyReLU(0.2),
            nn.LazyConv2d(512, kernel_size=4, stride=1, padding=1),
            nn.LazyBatchNorm2d(),
            nn.LeakyReLU(0.2),
            nn.LazyConv2d(1, kernel_size=4, stride=1, padding=1),
        ) 
    def forward(self, x, y):
        # Concatenate input and generated/real image
        out = torch.cat([x, y], dim=1)
        return self.model(out)

class Decoder(nn.Module):
    def __init__(self, out_channels=3, latent_dim=128):
        super(Decoder, self).__init__()
        self.feature_map_shape = (512, 4, 4)  # Starting shape after FC layer
        self.fc = nn.Linear(latent_dim, 512 * 4 * 4)

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 512, kernel_size=4, stride=2, padding=1),  # 4 → 8
            nn.ReLU(True),
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),  # 8 → 16
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 256, kernel_size=4, stride=2, padding=1),  # 16 → 32
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),  # 32 → 64
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),   # 64 → 128
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 64, kernel_size=4, stride=2, padding=1),    # 128 → 256
            nn.ReLU(True),
            nn.ConvTranspose2d(64, out_channels, kernel_size=4, stride=2, padding=1),  # 256 → 512
            nn.Sigmoid()  # or Tanh(), depending on your input normalization
        )

    def forward(self, z):
        x = self.fc(z)
        x = x.view(z.size(0), *self.feature_map_shape)
        x = self.decoder(x)
        return x

class ConvVAE(nn.Module):
    def __init__(self, in_channels=3, out_channels=1, latent_dim=128):
        super(ConvVAE, self).__init__()
        self.encoder = Encoder(in_channels=in_channels, latent_dim=latent_dim)
        self.decoder = Decoder(out_channels=out_channels, latent_dim=latent_dim)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        mu, logvar = self.encoder(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decoder(z)
        # print(f"recon shape: {recon.shape}, mu shape: {mu.shape}, logvar shape: {logvar.shape}")
        return recon, mu, logvar
class Decoder(nn.Module):
    def __init__(self, out_channels=3, latent_dim=128):
        super(Decoder, self).__init__()
        self.feature_map_shape = (512, 4, 4)  # Starting shape after FC layer
        self.fc = nn.Linear(latent_dim, 512 * 4 * 4)

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 512, kernel_size=4, stride=2, padding=1),  # 4 → 8
            nn.ReLU(True),
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),  # 8 → 16
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 256, kernel_size=4, stride=2, padding=1),  # 16 → 32
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),  # 32 → 64
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),   # 64 → 128
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 64, kernel_size=4, stride=2, padding=1),    # 128 → 256
            nn.ReLU(True),
            nn.ConvTranspose2d(64, out_channels, kernel_size=4, stride=2, padding=1),  # 256 → 512
            nn.Sigmoid()  # or Tanh(), depending on your input normalization
        )

    def forward(self, z):
        x = self.fc(z)
        x = x.view(z.size(0), *self.feature_map_shape)
        x = self.decoder(x)
        return x

class ConvVAE(nn.Module):
    def __init__(self, in_channels=3, out_channels=1, latent_dim=128):
        super(ConvVAE, self).__init__()
        self.encoder = Encoder(in_channels=in_channels, latent_dim=latent_dim)
        self.decoder = Decoder(out_channels=out_channels, latent_dim=latent_dim)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        mu, logvar = self.encoder(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decoder(z)
        # print(f"recon shape: {recon.shape}, mu shape: {mu.shape}, logvar shape: {logvar.shape}")
        return recon, mu, logvar
```

```python
import cv2
import numpy as np

def get_model_predictions_cv2(terrain_model, heightmap_model, dataloader, device, max_batches=1, index=0):
    """
    Returns input and output image batches from model evaluation (first N batches),
    converted to OpenCV BGR image arrays.

    Returns:
        List of tuples: [(input_cv2_image, output_cv2_image), ...]
    """
    # model.eval()
    terrain_model.eval()
    heightmap_model.eval()
    predictions = []

    with torch.no_grad():
        for count, (height, _, segmentation) in enumerate(dataloader):
            if(count < index):
                continue
            inputs = segmentation.to(device)

            # outputs = model(inputs)
            terrain_outputs = terrain_model(inputs)
            height_outputs = heightmap_model(inputs)
            # outputs = torch.sigmoid(outputs)  # if using BCE/MSE
            terrain_outputs = torch.sigmoid(terrain_outputs)
            height_outputs = torch.sigmoid(height_outputs)
            # height = height.to(device)  # Use heightmap as the correct image
            # Move tensors to CPU and convert to [H, W, C] NumPy format
            inputs_np = inputs.detach().cpu().permute(0, 2, 3, 1).clamp(0, 1).numpy()
            # outputs_np = outputs.detach().cpu().permute(0, 2, 3, 1).clamp(0, 1).numpy()
            # outputs_height_np = height_outputs.detach().cpu().permute(0, 2, 3, 1).clamp(0, 1).numpy()
            outputs_height_np = height_outputs.detach().cpu().permute(0, 2, 3, 1).numpy()
            outputs_terrain_np = terrain_outputs.detach().cpu().permute(0, 2, 3, 1).clamp(0, 1).numpy()
            # height_np = height.detach().cpu().permute(0, 2, 3, 1).clamp(0, 1).numpy()

            # Convert each image to OpenCV (BGR, uint8)
            for inp, out_height, out_terrain in zip(inputs_np, outputs_height_np, outputs_terrain_np):
                inp_bgr = cv2.cvtColor((inp * 255).astype(np.uint8), cv2.COLOR_RGB2BGR)
                out_height_bgr = cv2.cvtColor((out_height * 255).astype(np.uint8), cv2.COLOR_RGB2BGR)
                out_terrain_bgr = cv2.cvtColor((out_terrain * 255).astype(np.uint8), cv2.COLOR_RGB2BGR)
                # height_bgr = cv2.cvtColor((height_np * 255).astype(np.uint8), cv2.COLOR_RGB2BGR)
                # predictions.append((inp_bgr, out_bgr, height_bgr))
                predictions.append((inp_bgr, out_height_bgr, out_terrain_bgr))

            if max_batches is not None:
                max_batches -= 1
                if max_batches <= 0:
                    break

    return predictions
```
